```python
import requests
from bs4 import BeautifulSoup
import csv

# Modified function to filter by author
def scrape_einstein_quotes(soup, quotes):
    for quote in soup.find_all('div', class_='quote'):
        author = quote.find('small', class_='author').text
        
        # Check if the author is Albert Einstein
        if author == "Albert Einstein":
            text = quote.find('span', class_='text').text
            tags = ', '.join(tag.text for tag in quote.find_all('a', class_='tag'))
            quotes.append({'Text': text, 'Author': author, 'Tags': tags})

base_url = 'https://quotes.toscrape.com'
headers = {'User-Agent': 'Mozilla/5.0'}
quotes = []

def scrape_all_pages(url):
    while url:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')
        scrape_einstein_quotes(soup, quotes)
        
        next_page = soup.find('li', class_='next')
        url = base_url + next_page.find('a')['href'] if next_page else None

# Execution
scrape_all_pages(base_url)

# Save to quotes_Einstein.csv
with open('quotes_Einstein.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=['Text', 'Author', 'Tags'])
    writer.writeheader()
    writer.writerows(quotes)

print("Albert Einstein quotes have been saved to quotes_Einstein.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

def scrape_life_quotes(soup, quotes):
    for quote in soup.find_all('div', class_='quote'):
        text = quote.find('span', class_='text').text
        
        # Check if "life" is in the text (case-insensitive check is often safer)
        if "life" in text.lower():
            author = quote.find('small', class_='author').text
            tags = ', '.join(tag.text for tag in quote.find_all('a', class_='tag'))
            quotes.append({'Text': text, 'Author': author, 'Tags': tags})

base_url = 'https://quotes.toscrape.com'
headers = {'User-Agent': 'Mozilla/5.0'}
quotes = []

def scrape_all_pages(url):
    while url:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')
        scrape_life_quotes(soup, quotes)
        
        next_page = soup.find('li', class_='next')
        url = base_url + next_page.find('a')['href'] if next_page else None

# Execution
scrape_all_pages(base_url)

# Save to life_quotes.csv
with open('life_quotes.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=['Text', 'Author', 'Tags'])
    writer.writeheader()
    writer.writerows(quotes)

print("Quotes containing 'life' have been saved to life_quotes.csv")
```

```python
# Name: Nik Amir Faris Bin Nik Amidi 
# Name: Muhammad Thaqif Bin KhairulAnuar
# Student ID: SW01083709 & SW01083763



import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# Function to get page HTML
def get_page_content(url):
   def get_page_content(url):
    
    headers = {
        "User-Agent": "Mozilla/5.0",
        "Accept-Language": "en-US,en;q=0.9"
    }
    
    response = requests.get(
        url,
        headers=headers,
        proxies={"http": None, "https": None}
    )
    
    if response.status_code == 200:
        return response.text
    else:
        print("Failed to retrieve page:", response.status_code)
        return None


# Function to extract reviews
def extract_reviews(html):
    """
    Extract reviewer name,
    review date,
    review content
    from Shopee page.
    """
    
    soup = BeautifulSoup(html, "html.parser")
    reviews = []
    
    review_blocks = soup.find_all("div", class_="item")
    
    for review in review_blocks:
        
        name = review.find("span", class_="middle")
        name = name.text.strip() if name else "N/A"
        
        date = review.find("span", class_="lite")
        date = date.text.strip() if date else "N/A"
        
        content = review.find("div", class_="content")
        content = content.text.strip() if content else "N/A"
        
        reviews.append([name, date, content])
        
    return reviews


# MAIN PROGRAM
all_reviews = []

base_url = "https://shopee.com.my/Apple-iPhone-17-Pro-i.1278832578.44264342781?extraParams=%7B%22display_model_id%22%3A208656041231%2C%22model_selection_logic%22%3A3%7D&sp_atk=ab6adf9c-e8c7-48a1-af32-b2f6904bf5c9&xptdk=ab6adf9c-e8c7-48a1-af32-b2f6904bf5c9"

for page in range(1, 6):
    
    print(f"Scraping page {page}...")
    
    url = base_url + f"&page={page}"
    
    html = get_page_content(url)
    
    
    if html:
        page_reviews = extract_reviews(html)
        all_reviews.extend(page_reviews)
    
    time.sleep(2)


df = pd.DataFrame(all_reviews, columns=["Reviewer Name", "Review Date", "Review Content"])
df.to_csv("Shopee_reviews.csv", index=False)

print("Scraping completed successfully.")
```
